```python
import socket
import subprocess
import platform

def ping_host(host):
    print(f"\nPinging {host}...")
    try:
        param = "-n" if platform.system().lower() == "windows" else "-c"
        output = subprocess.run(["ping", param, "4", host], capture_output=True, text=True)
        print(output.stdout)
    except Exception as e:
        print("Ping failed:", e)

def dns_lookup(domain):
    print(f"\nLooking up DNS for {domain}...")
    try:
        ip = socket.gethostbyname(domain)
        print(f"IP Address: {ip}")
    except socket.gaierror:
        print("DNS lookup failed.")

def scan_port(host, port):
    print(f"\nChecking port {port} on {host}...")
    try:
        s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        s.settimeout(2)
        result = s.connect_ex((host, port))
        if result == 0:
            print(f"Port {port} is OPEN")
        else:
            print(f"Port {port} is CLOSED")
        s.close()
    except Exception as e:
        print("Port scan failed:", e)

def get_local_ip():
    hostname = socket.gethostname()
    local_ip = socket.gethostbyname(hostname)
    print(f"\nLocal IP Address: {local_ip}")

def traceroute(host):
    print(f"\nRunning traceroute for {host}...\n")
    try:
        system = platform.system().lower()

        if system == "windows":
            cmd = ["tracert", host]
        else:
            cmd = ["traceroute", host]

        output = subprocess.run(cmd, capture_output=True, text=True)
        print(output.stdout)

    except Exception as e:
        print("Traceroute failed:", e)

def main():
    while True:
        print("\n--- Network Troubleshooter ---")
        print("1. Ping Host")
        print("2. DNS Lookup")
        print("3. Scan Port")
        print("4. Get Local IP")
        print("5. Traceroute")
        print("6. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            host = input("Enter host/IP: ")
            ping_host(host)

        elif choice == "2":
            domain = input("Enter domain: ")
            dns_lookup(domain)

        elif choice == "3":
            host = input("Enter host/IP: ")
            port = int(input("Enter port: "))
            scan_port(host, port)

        elif choice == "4":
            get_local_ip()

        elif choice == "5":
            host = input("Enter host/IP: ")
            traceroute(host)

        elif choice == "6":
            print("Exiting...")
            break

        else:
            print("Invalid choice.")

if __name__ == "__main__":
    main()
```
